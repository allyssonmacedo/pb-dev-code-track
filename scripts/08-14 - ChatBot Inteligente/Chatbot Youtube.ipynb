### Criando um chatbot Inteligente

```python
 #Criar um ambiente virtual no anaconda ; Abrir o prompt do anaconda:
#conda create --name chatbotpython python=3.6
#conda deactivate
#pip install jupyter
#pip install chatterbot
#pip install spacy -------> caso o chatterbot não instale isso
#Abrir o Jupyter no ambiente virtual:
#jupyter notebook

from chatterbot import ChatBot
from chatterbot.trainers import ListTrainer

# isso aqui só precisa para corrigir o bug
from spacy.cli import download

download("en_core_web_sm")

class ENGSM:
    ISO_639_1 = 'en_core_web_sm'
```

```python
chatbot = ChatBot("BotLira", tagger_language=ENGSM)

conversa = [
    "Coe",
    "E aí, tranquilo?",
    "Tranquilo",
    "Qual a boa de hoje?",
    "a Hashtag tá ensinando Python e até chatbot",
    "Caraca que doidera",
    "Maneiro né",
    "Irado",
]

trainer = ListTrainer(chatbot)
trainer.train(conversa)
```

```python
while True:
    mensagem = input("Mande uma mensagem para o chatbot:")
    if mensagem == "parar":
        break
    resposta = chatbot.get_response(mensagem)
    print(resposta)
```
